```python
delivered_list["cohort"] = delivered_list.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period("M")
delivered_list
```

```python
delivered_list_c = delivered_list.query("'2017-01' <= cohort <= '2017-12'") \
    .groupby(["cohort", "order_purchase_timestamp"], as_index=False) \
    .agg({"customer_unique_id": "nunique"})\
    .rename(columns=({'cohort':'first_purchase'}))
delivered_list_c
```

```python
delivered_list_c["order_purchase_timestamp"] = pd.to_datetime(delivered_list_c["order_purchase_timestamp"]).dt.to_period('M')
```

```python
delivered_list_c['period'] = (delivered_list_c.order_purchase_timestamp - delivered_list_c.first_purchase).apply(attrgetter('n'))
delivered_list_c.info()
```

```python
delivered_list_cohort = delivered_list_c.groupby(['first_purchase','period'],as_index=False).agg({'customer_unique_id':'count'})\
        .pivot_table(index='first_purchase', columns = 'period', values = 'customer_unique_id').fillna(0)
delivered_list_cohort
```

```python
delivered_list_cohort=round(delivered_list_cohort.div(delivered_list_cohort.iloc[:,0], axis = 0).mul(100),3)
```

```python
last_day = delivered_list.order_purchase_timestamp.max()
```

```python
RFM = delivered_list.groupby('customer_unique_id') \
    .agg({'order_purchase_timestamp': lambda x: (last_day - x.max()).days, 'product_id': lambda x: len(x), 'price': lambda x: x.sum()}) \
    .rename(columns={'order_purchase_timestamp': 'R', 'product_id': 'F', 'price': 'M'})
RFM
```

```python
#Разобьем на диапазон от 1 до 5: 0-20, 20-40, 40-60, 60-80, 80-100%
```

```python
seg = RFM.quantile(q=[0.20, 0.40, 0.60, 0.80]).to_dict()
seg
```

```python
#Сегментируем всех покупателей с помощью функции
def customer_seg (x,y,z):
    if x <= z[y][0.20]:
        return 1
    elif x <= z[y][0.40]:
        return 2
    elif x <= z[y][0.60]:
        return 3
    elif x <= z[y][0.80]:
        return 4
    else:
        return 5
```

```python
RFM['R_seg'] = RFM['R'].apply(customer_seg, args=('R', seg, ))
RFM['F_seg'] = RFM['F'].apply(customer_seg, args=('F', seg, ))
RFM['M_seg'] = RFM['M'].apply(customer_seg, args=('M', seg, ))
```

```python
RFM['RFM_overall'] = RFM.R_seg.map(str) + RFM.F_seg.map(str) + RFM.M_seg.map(str)
RFM
```
